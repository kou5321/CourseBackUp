```python
import pandas as pd
data = pd.read_excel('./bilibili.xlsx')
print(data.shape)
data.info()
```

# 删除掉了一些nan数据和没什么用的数据

```python
data.drop(['badge'],axis=1, inplace=True)
data.drop(['badge_type'],axis=1, inplace=True)
data.drop(['tags'],axis=1, inplace=True)
# data.drop(['score'],axis=1, inplace=True)
data.drop(['pub_date'],axis=1, inplace=True)
data.drop(['renewal_time'],axis=1, inplace=True)
data.drop(['season_id'],axis=1, inplace=True)
data.drop(['media_id'],axis=1, inplace=True)
data.drop(['pub_real_time'],axis=1, inplace=True)

# 删除含有NaN的数据
data.dropna(inplace=True)
data.info()
```

# 整理了follow的数据，改成了数字

```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='./STSONG.TTF',size=10)
# df = pd.read_excel('/data/course_data/data_analysis/lagou.xlsx')
data_is_finish = data.is_finish
# data.info
plt.figure(figsize = (10,8))
num = data.groupby('is_finish').size()
print(num)
# 绘制饼图
patches, l_text, p_text = plt.pie(num,labels = num.index, autopct='%.1f%%',shadow=False,startangle=90)
for t in l_text: 
    t.set_fontproperties(my_font)
plt.axis('equal')
plt.legend(prop = my_font)
plt.title('bilibili番剧完结比例',fontproperties=my_font)
plt.savefig('完结比例.png')
plt.show()
```
